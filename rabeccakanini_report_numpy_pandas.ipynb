```python
# Line PLot

import matplotlib.pyplot as plt

x = [1,2,3,4]

y = [3,6,9,12]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```
